```python
#1 Read the Text File From my Computer
filename = "sample.txt"
with open(filename, "r") as f:
    print(f.read())
```

```python
#2 Read the first n lines of a text file
filename = "sample.txt"
def read_first_n_lines(filename, n):
    with open(filename, "r") as f:
        lines = f.readlines()
        return lines[:n]
print(read_first_n_lines(filename, 2))
```

```python
#3 Append text to a file and display the text file's contents
filename = "sample.txt"
def append_and_display(filename, text):
    with open(filename, "a") as f:
       f.write(text + "\n")
    with open(filename, "r") as f:
        return f.read()
print(append_and_display(filename, "This is a new appended text line"))
```

```python
#4 Ask the user for n-line and read last n lines of a file, display
filename = "sample.txt"
def read_last_of_n_lines(filename, n):
    with open(filename, "r") as f:
        lines = f.readlines()
    return [line.strip() for line in lines[-n:]]
print(read_last_of_n_lines(filename, 2))
```

```python
#5 Read a file line by line and store it into a list. Display the list.
filename = "sample.txt"
def read_the_file_to_list(filename):
  with open (filename, "r") as f:
    lines = f.readlines()
    return [line.strip() for line in lines]
print(read_the_file_to_list(filename))
```

```python
#6 Read a file line by line store it into a variable. Display the variable.
filename = "sample.txt"
def read_the_file_into_a_variable(filename):
  with open (filename, "r") as f:
   return "".join(f.readlines())
print(read_the_file_into_a_variable(filename))
```

```python
#7 Find the longest words in a text file. Display the longest word.
filename = "sample.txt"
def longest_word_finder(filename):
  with open(filename, "r") as f:
    words = f.read().split()
    longest_word = max(words, key=len)
    return longest_word
print(longest_word_finder(filename))
```

```python
#8 Count the number of lines in a text file. Display the number of lines.
filename = "sample.txt"
def line_counter(filename):
  with open(filename, "r") as f:
    lines = f.readlines()
    return len(lines)
print(line_counter(filename))
```
